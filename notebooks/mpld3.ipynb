# MPLD3
[mpld3](https://mpld3.github.io) conceptually works as a backend for matplotlib. You make your figures with matplotlib commands that you probably know already and then save it as a html website with mpld3. 

`mpld3` is *stable* - it's not under active development and no new features are added, but the maintainers update it regularly to make sure it keeps working with new version of numpy ans matplotlib.

If you already know matplotlib, `mpld3` is a great way to get started with interactive plots; though not all (but most) matplotlib features are supported.

```python
import matplotlib.pyplot as plt
import numpy as np
import mpld3
from mpld3 import plugins
```

Here, I'm using a table with data from one of my old papers. We want to compare optical reddening and X-ray absorption. There are going to be a lot of stars and sometimes we may want to know more about each star, but we also don't have the space to label each and every point in our plot or the names would all overlap. This is a great use of an interactive figure! Just use your mouse to identify the stars you're interested in.

```python
gs08tab = np.genfromtxt('../data/guenther_schmitt_2008.tab', delimiter=',', names=True,
              dtype=[float, float, float, float, 'U10', float, float, float])
```

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(111)

ax.set_yscale("log")
ax.set_xscale("log")




points = ax.plot(gs08tab['av'], gs08tab['nh'] * 1e21, 
            'rs', label='GS08')
# AA Tau - Grosso et al 2007
ax.plot([0.78] * 9, np.array([0.95, 1.54, 1.80, 1.36, 1.33, 1.04, 1.54, 1.02, 1.53]) * 1e22, 
'.', color='orange', label='AA Tau')
ax.plot(np.array([4]), np.array([0.6e22]), 'o', color='orange', label='_AA Tau extra abs')
# TWA 30 - Principe et al 2016
ax.plot([14.9, 4.7], [9.5e21, 1.6e21], 'bp', label='TWA 30A')

x = np.arange(*ax.get_xlim(),step = .01)
ax.plot(x, x * 1.8e21, 'g', label='ISM')
#ax.plot(x, x*0.89 * 0.112 * 1e22, 'k--', label='NGC1333')
ax.plot(x, x*0.63 * 1e22 *0.112, 'g:', label='Serpens')
ax.plot(x, x*1.4 * 0.112 * 1e22, 'g--', label=r'$\rho$ Oph')
ax.set_xlabel('$A_V$ [mag]')
ax.set_ylabel('$N_H$ [cm$^{-2}$]')
ax.set_ylim([1e19, 3e24])
ax.set_xlim([.005, None])  # cuts of some low A_V sources from GS08
# Manually reorder the legend
legend_order=[4, 5, 0, 1, 2, 3]
handles, labels = ax.get_legend_handles_labels()

handles = [handles[i] for i in legend_order]
labels = [labels[i] for i in legend_order]
ax.legend(handles, labels, loc='upper left', numpoints=1, ncol=2, fontsize='small')
```

Above, the matplotlib plot. Below, the mpld3 plot - essentially identical code except to make the tooltips that matplotlib does not have!

```python
fig = plt.figure(figsize=(4,3))
ax = fig.add_subplot(111)

ax.set_yscale("log")
ax.set_xscale("log")


points = ax.plot(gs08tab['av'], gs08tab['nh'] * 1e21, 
            'rs', label='GS08')
# AA Tau - Grosso et al 2007
ax.plot([0.78] * 9, np.array([0.95, 1.54, 1.80, 1.36, 1.33, 1.04, 1.54, 1.02, 1.53]) * 1e22, 
'.', color='orange', label='AA Tau')
ax.plot(np.array([4]), np.array([0.6e22]), 'o', color='orange', label='_AA Tau extra abs')
# TWA 30 - Principe et al 2016
ax.plot([14.9, 4.7], [9.5e21, 1.6e21], 'bp', label='TWA 30A')

x = np.arange(*ax.get_xlim(),step = .01)
ax.plot(x, x * 1.8e21, 'g', label='ISM')
#ax.plot(x, x*0.89 * 0.112 * 1e22, 'k--', label='NGC1333')
ax.plot(x, x*0.63 * 1e22 *0.112, 'g:', label='Serpens')
ax.plot(x, x*1.4 * 0.112 * 1e22, 'g--', label=r'$\rho$ Oph')
ax.set_xlabel('$A_V$ [mag]')
ax.set_ylabel('$N_H$ [cm$^{-2}$]')
ax.set_ylim([1e19, 3e24])
ax.set_xlim([.005, None])  # cuts of some low A_V sources from GS08
# Manually reorder the legend
legend_order=[4, 5, 0, 1, 2, 3]
handles, labels = ax.get_legend_handles_labels()

handles = [handles[i] for i in legend_order]
labels = [labels[i] for i in legend_order]
ax.legend(handles, labels, loc='upper left', numpoints=1, ncol=2, fontsize='small')


#labels = []
#for i in range(N):
#    label = df.iloc[[i], :].T
#    label.columns = ['Row {0}'.format(i)]
#    # .to_html() is unicode; so make leading 'u' go away with str()
#    labels.append(str(label.to_html()))
tooltiplabels = list(gs08tab['name'])

tooltip = plugins.PointHTMLTooltip(points[0], tooltiplabels,
                                   voffset=10, hoffset=10)
plugins.connect(fig, tooltip)

mpld3.display()
```
